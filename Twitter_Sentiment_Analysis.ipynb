# Sentiment Analysis on Twitter Dataset

# Objective

- To analyze sentiments from Twitter text data.
- To visualize positive, negative, and neutral sentiments.
- To perform data cleaning and preprocessing.
- To generate meaningful insights using charts and graphs.
- To understand public opinions and social media trends.

```python
# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset

df = pd.read_csv("Twitter Sentiment Dataset.csv")

# Display First 5 Rows

df.head()
```

```python
# Shape of Dataset

df.shape
```

```python
# columns Name

df.columns
```

```python
# Dataset Information

df.info()
```

```python
# Statistical Summary

df.describe()
```

```python
# Checking Missing Values

df.isnull().sum()
```

```python
# Removing Missing Values

df.dropna(inplace = True)
```

```python
# Checking Duplicate Rows

df.duplicated().sum()
```

```python
# Removing Duplicate Rows

df.drop_duplicates(inplace = True)
```

```python
# Checking Unique Sentiments

df['category'].unique()
```

```python
# Sentiment Distribution

sns.countplot(x='category', data=df)

plt.title("Sentiment Distribution")

plt.xlabel("Sentiment")

plt.ylabel("Count")

plt.show()
```

```python
# Sentiment Percentage

sentiment_counts = df['category'].value_counts()

plt.pie(sentiment_counts,
        labels=['Negative', 'Neutral', 'Positive'],
        autopct='%1.1f%%')

plt.title("Sentiment Percentage")

plt.show()
```

```python
# Creating Tweet Length Column

df['tweet_length'] = df['clean_text'].apply(len)

# Displaying text with tweet length

df[['clean_text', 'tweet_length']].head()
```

```python
# Distribution of Tweet Length

sns.histplot(df['tweet_length'], bins=30)

plt.title("Tweet Length Distribution")

plt.xlabel("Tweet Length")

plt.ylabel("Frequency")

plt.show()
```

```python
# Average Tweet Length by Sentiment

avg_length = df.groupby('category')['tweet_length'].mean()

avg_length
```

```python
# Barplot of Average Tweet Length

avg_length.plot(kind='bar')

plt.title("Average Tweet Length by Sentiment")

plt.xlabel("Sentiment")

plt.ylabel("Average Length")

plt.show()
```

```python
# Most Frequent Sentiment

df['category'].mode()
```

```python
#Saving Cleaned Dataset

df.to_csv("cleaned_twitter_sentiment_data.csv", index=False)
```
