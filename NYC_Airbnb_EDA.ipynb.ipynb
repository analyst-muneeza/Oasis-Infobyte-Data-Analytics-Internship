# Data Cleaning on New York City Airbnb Open Data 

# Objective
- To perform data cleaning on the New York City Airbnb Open Dataset.
- To identify and handle missing values, duplicate records, and inconsistent data.
- To detect and remove outliers from the dataset.
- To convert incorrect data types into proper formats.
- To improve the overall quality and reliability of the dataset for future analysis.

```python
# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset
df = pd.read_csv("New York City Airbnb Open Data.csv")

# Display First 5 Rows
df.head()
```

```python
# Shape of Dataset

df.shape
```

```python
# columns Name

df.columns
```

```python
# Dataset Information

df.info()
```

```python
# Statistical Summary

df.describe()
```

```python
# checking missing values

df.isnull().sum()
```

```python
# Handling Missing Values

# Filling missing values in reviews_per_month
df['reviews_per_month'] = df['reviews_per_month'].fillna(
    df['reviews_per_month'].median()
)

# Filling missing values in text columns
df['name'] = df['name'].fillna("Unknown")

df['host_name'] = df['host_name'].fillna("Unknown")
```

```python
# Removing Remaining Null Values

df.dropna(subset=['last_review'], inplace=True)
```

```python
# Checking Duplicate Rows

df.duplicated().sum()
```

```python
# Removing Duplicate Rows

df.drop_duplicates(inplace=True)
```

```python
# Converting Data Types

df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# Checking Data Types

df.dtypes
```

```python
# Standardizing Column Names

df.columns = df.columns.str.lower()
```

```python
# Removing Extra Spaces

df['neighbourhood_group'] = df['neighbourhood_group'].str.strip()
```

```python
# Detecting Outliers

plt.figure(figsize=(10,5))

sns.boxplot(x=df['price'],
            color='skyblue')

plt.title("Boxplot of Price")
plt.xlabel("Price")

plt.show()
```

```python
# Removing Outliers

df = df[df['price'] < 1000]
```

```python
# Heatmap of Missing Values

plt.figure(figsize=(10,6))

sns.heatmap(df.isnull(),
            cbar=False,
            cmap='viridis')

plt.title("Missing Values Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")

plt.show()
```

```python
# Room Type Distribution

sns.countplot(x='room_type', data=df)
plt.title("Room Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Price Distribution

sns.histplot(df['price'], bins=50)
plt.title("Price Distribution")
plt.show()
```

```python
# Saving Cleaned Dataset

df.to_csv("cleaned_airbnb_data.csv", index=False)
```
