# EDA on McDonald’s Nutrition Dataset

# Objective

- To perform Exploratory Data Analysis (EDA) on McDonald’s nutrition dataset.
- To analyze calories, fat, sugar, protein, and sodium values.
-To identify healthy and unhealthy food items.
- To visualize nutritional patterns using charts and graphs.
- To generate meaningful insights from the dataset.

```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv("McDonalds_Nutrition_Dataset.csv")

# Display first 5 rows
df.head()
```

```python
# Checking dataset information
df.info()
```

```python
# Checking missing values
df.isnull().sum()
```

```python
# Checking duplicate rows
df.duplicated().sum()
```

```python
# Statistical summary of dataset
df.describe()
```

```python
# Display column names
df.columns
```

```python
# Finding item with highest calories
highest_calorie = df.loc[df['Calories'].idxmax()]

print(highest_calorie)
```

```python
# Finding item with highest sugar
highest_sugar = df.loc[df['Sugars'].idxmax()]

print(highest_sugar)
```

```python
# Average calories by category
avg_calories = df.groupby('Category')['Calories'].mean()

print(avg_calories)
```

```python
# Top 10 high calorie items
top10 = df.sort_values(by='Calories', ascending=False).head(10)

print(top10[['Item', 'Calories']])
```

```python
# Bar chart for top 10 high calorie items

plt.figure(figsize=(10,6))

sns.barplot(x='Calories', y='Item', data=top10)

plt.title("Top 10 High Calorie McDonald's Items")

plt.show()
```

```python
# Scatter plot between Calories and Total Fat

plt.figure(figsize=(8,6))

sns.scatterplot(x='Total Fat', y='Calories', data=df)

plt.title("Calories vs Total Fat")

plt.show()
```

```python
# Heatmap for correlation

plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Histogram of calories distribution

plt.figure(figsize=(8,6))

sns.histplot(df['Calories'], bins=20)

plt.title("Calories Distribution")

plt.show()
```

# Conclusion insights

1. High fat items usually contain high calories.
2. Some beverages contain very high sugar.
3. Salads are healthier than burgers and fries.
4. Nutrition values vary significantly across categories.
