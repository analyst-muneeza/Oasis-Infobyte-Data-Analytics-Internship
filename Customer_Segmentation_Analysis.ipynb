```python
# Plotting education graduation distribution

plt.figure(figsize=(6,5))

sns.countplot(x='education_Graduation', data=df)

plt.title("Education Graduation Distribution")

plt.show()
```

```python
# Plotting income distribution

plt.figure(figsize=(8,5))

sns.histplot(df['Income'], bins=20)

plt.title("Income Distribution")

plt.show()
```

```python
# Scatter plot between income and total spending

plt.figure(figsize=(8,6))

sns.scatterplot(
    x='Income',
    y='Total_Spending',
    data=df
)

plt.title("Income vs Total Spending")

plt.show()
```
